## Import Sodium's YOLO or Solo

```
import sodium.yolo as solo
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
yolo = solo.YoloOpenCV()
```

## Download the images

```
import urllib.request
print('Beginning file download with urllib2...')
BASE_URL = 'https://github.com/Enamulla577/EVAI/tree/main/Session13/Images/'

myfile = 'person.jpg'
urllib.request.urlretrieve(BASE_URL + myfile, myfile)
```

```
img = yolo.run_yolo_on_image('person.jpg')
plt.figure(figsize=(10, 15))
plt.imshow(img)
plt.axis('off')
plt.show()
```
